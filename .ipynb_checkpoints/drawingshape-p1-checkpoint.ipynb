# Recreate the OpenCV logo using OpenCV drawing functions only


You must draw:
- Three colored shapes (ellipse: Blue, Green, Red) arranged in a triangular pattern.
- Proper positioning and size of circles.
- The text "OpenCV" at the center/bottom.

## Explanation


Let the image canvas have width `w` and height `h`.  
We’ll position the logo so that the centers of the three circles form an **equilateral triangle** centered on the canvas.
To make it easy, we can assign the edge of the triangle equals to `w//2`
$$AB = AC = BC = w//2$$

### Find the radius of all there elilipses
As we can see, **the radius of all three ellipses is ` x `**

#### Relationship between height and radius
The total canvas height $h$ equals twice the circle radius plus the remaining top gap $AH$.
Therefore:

$$
h = 2x + AH
$$

$$
\Rightarrow x = \frac{h - AH}{2}
$$

Since $AH = \cos(30^\circ) \times \frac{w}{2}$ ($AH$ is the vertical component of the line from the top vertex to the midpoint of the base in the equilateral triangle formed by the three circle centers), we can substitute:

$$
x = \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2}
$$


### Finding the Centers of the Ellipses (A, B, C)

In **OpenCV**, the coordinate system starts from the **top-left corner** of the image,  
where the **x-axis** increases to the right and the **y-axis** increases downward.

The centers of the three ellipses can therefore be defined as:

$$
A\left(\frac{w}{2},\ x\right)
  = A\left(\frac{w}{2},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2}\right)
$$

$$
B\left(\frac{w}{2} - \frac{w}{4},\ x + \cos(30^\circ) \times \frac{w}{2}\right)
  = B\left(\frac{w}{4},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2} + \cos(30^\circ) \times \frac{w}{2}\right)
$$

$$
C\left(\frac{w}{2} + \frac{w}{4},\ x + \cos(30^\circ) \times \frac{w}{2}\right)
  = C\left(\frac{3w}{4},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2} + \cos(30^\circ) \times \frac{w}{2}\right)
$$


Here:
- $A$ represents the **top ellipse** (red),
- $B$ the **bottom-left ellipse** (green),
- $C$ the **bottom-right ellipse** (blue).  
The variable $x$ is the circle radius derived earlier, and $AH$ is the vertical offset determined by the triangle geometry.


### Understanding the `gap` Variable

In OpenCV drawing, each ellipse is positioned based on its **center coordinates**.  
However, if we place the circles using pure geometry alone, they often **touch or overlap too much**, especially when converting continuous math coordinates into pixel-based drawing.

To fix that, we introduce a small vertical and horizontal **offset** called `gap`.

The `gap` variable adds spacing to visually balance the three ellipses:
- It slightly **lowers the bottom ellipses** (green and blue).
- It also **shifts each circle horizontally**, preventing them from touching too tightly.

So finally, the centers of the ellipses are:

$$
A\left(\frac{w}{2} + \text{gap},\ x\right)
  = A\left(\frac{w}{2} + \text{gap},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2}\right)
$$

$$
B\left(\frac{w}{2} - \frac{w}{4} - \frac{\text{gap}}{2} + \text{gap},\ x + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
  = B\left(\frac{w}{2} - \frac{w}{4} - \frac{\text{gap}}{2} + \text{gap},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2} + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
$$

$$
C\left(\frac{w}{2} + \frac{w}{4} + \frac{\text{gap}}{2} + \text{gap},\ x + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
  = C\left(\frac{w}{2} + \frac{w}{4} + \frac{\text{gap}}{2} + \text{gap},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2} + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
$$

## Code Demo

### Import all essential library

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Create a canvas 1400 x 1200 pixels

```python
h, w = 1400,1200
img = np.full((h, w, 3), 255, dtype=np.uint8)
```

### Define hyperparemeter such as center position and radius of the ellipse

$$
A\left(\frac{w}{2} + \text{gap},\ x\right)
  = A\left(\frac{w}{2} + \text{gap},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2}\right)
$$

$$
B\left(\frac{w}{2} - \frac{w}{4} - \frac{\text{gap}}{2} + \text{gap},\ x + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
  = B\left(\frac{w}{2} - \frac{w}{4} - \frac{\text{gap}}{2} + \text{gap},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2} + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
$$

$$
C\left(\frac{w}{2} + \frac{w}{4} + \frac{\text{gap}}{2} + \text{gap},\ x + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
  = C\left(\frac{w}{2} + \frac{w}{4} + \frac{\text{gap}}{2} + \text{gap},\ \frac{h - \big(\cos(30^\circ) \times \frac{w}{2}\big)}{2} + \cos(30^\circ) \times \frac{w}{2} + \text{gap}\right)
$$

```python
h_logo, w_logo = 900,900
gap = 30
r = (h_logo - (np.sqrt(3)/2)*(w_logo/2))/2 - gap

#A
center_red = (w_logo//2 + gap,int(r))

#B
center_green = ((w_logo//4) - gap//2 + gap, int(r + (np.sqrt(3)/2)*(w_logo/2)) + gap)

#C
center_blue = ((w_logo//2) + (w_logo//4) + gap//2 + gap, int(r + (np.sqrt(3)/2)*(w_logo/2)) + gap)
```

### Draw ellipse

#### Red Ellipse (A)
- **Outer ellipse:** red color `(0, 0, 255)` → drawn filled (`thickness = -1`).
- **Inner ellipse:** white `(255, 255, 255)` → smaller radius → creates a hollow center.
- **Angle = 120°:** rotates the ellipse so that the 60° open “cut” faces the correct direction.
- **Arc (0° → 300°):** draws 300° of the circle, leaving a 60° gap.

#### Green Ellipse (B)
- **Outer ellipse:** green color `(0, 255, 0)` → drawn filled (`thickness = -1`).
- **Inner ellipse:** white `(255, 255, 255)` → smaller radius → creates a hollow center.
- **Angle = 120°:** rotates the ellipse so that the 60° open “cut” faces the correct direction.
- **Arc (0° → 300°):** draws 300° of the circle, leaving a 60° gap.

#### Blue Ellipse (C)
- **Outer ellipse:** blue color `(255, 0, 0)` → drawn filled (`thickness = -1`).
- **Inner ellipse:** white `(255, 255, 255)` → smaller radius → creates a hollow center.
- **Angle = 120°:** rotates the ellipse so that the 60° open “cut” faces the correct direction.
- **Arc (0° → 300°):** draws 300° of the circle, leaving a 60° gap.

```python
#Red Ellipse
cv2.ellipse(img,center_red, (int(r),int(r)), 120, 0, 300, (0,0,255), -1)
cv2.ellipse(img,center_red, (int(r)//2,int(r)//2), 120, 0, 300, (255,255,255), -1)

#Green Ellipse
cv2.ellipse(img,center_green,(int(r),int(r)), 360, 0, 300, (0,255,0),-1)
cv2.ellipse(img,center_green,(int(r)//2,int(r)//2), 360, 0, 300, (255,255,255),-1)

#Blue Ellipse
cv2.ellipse(img,center_blue,(int(r),int(r)), 300, 0, 300, (255,0,0),-1)
cv2.ellipse(img,center_blue,(int(r)//2,int(r)//2), 300, 0, 300, (255,255,255),-1);
```

### Put text in the image
Finally, we add the text label **"OpenCV"** below the logo using the `cv2.putText()` function.


```python
font = cv2.FONT_HERSHEY_SIMPLEX
text = "OpenCV"
font_scale = 8
thickness = 30

cv2.putText(img, text, (0, 1100), font, font_scale, (0, 0, 0), thickness, cv2.LINE_AA);
```

```python
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
